```python
from matplotlib import pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 9, 6 # Define as dimensões dos gráficos
```

```python
nome = 'Calabresa', 'Carne seca', 'Carbonara', 'Napolitana'
valores = [2, 2, 2, 2]
destaque = (0, 0, 0, 0)
```

```python
plt.pie(valores, explode=destaque, labels=nome, autopct='%1.2f%%', shadow=True, startangle=90)
plt.legend(nome, title='Sabores', loc='right')
plt.axis('equal') # Permite que a legenda se movimente conforme o tamanho da figura
plt.title('Pizzas')
plt.show()
```

```python
nome = 'Carlos', 'Ana', 'Paulo', 'Lucas', 'Daniel', 'Bianca', 'João'
valores = [100, 90, 80, 70, 200, 250, 400]
destaque = (0, 0, 0, 0, 0, 0, 0.1)
```

```python
plt.pie(valores, explode=destaque, labels=nome, autopct='%1.2f%%', shadow=True, startangle=90)
plt.legend(nome, title='Apostadores', loc='best')
plt.axis('equal') # Permite que a legenda se movimente conforme o tamanho da figura
plt.title('Aposta Mega da virada')
plt.show()
```

```python
grupo = {'Carlos':100, 'Ana':90, 'Paulo':80, 'Lucas':70, 'Daniel':200, 'Bianca':250, 'João':400}
nome = tuple(grupo.keys())
valores = list(grupo.values())
```

```python
plt.pie(valores, explode=destaque, labels=nome, autopct='%1.2f%%', shadow=True, startangle=90)
plt.legend(nome, title='Apostadores', loc='best')
plt.axis('equal') # Permite que a legenda se movimente conforme o tamanho da figura
plt.title('Aposta Mega da virada')
plt.show()
```
