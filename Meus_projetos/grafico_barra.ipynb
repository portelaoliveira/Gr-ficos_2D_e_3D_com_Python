```python
from matplotlib import pyplot as plt
from numpy import*
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 6 # Define as dimensões dos gráficos
```

```python
def legenda_automatica(grafico):
    """ Adiciona a legenda acima da coluna do gráfico de acordo com a altura desta """
    for i in grafico:
        comprimento = i.get_width()
        plt.annotate(f'{comprimento}', xy=(comprimento, i.get_y() + i.get_height()/2), xytext=(10, 0), textcoords="offset points", ha="center")
```

```python
legenda = ["Aluno1", "Aluno2", "Aluno3", "Aluno4", "Aluno5"]
notas_1 = [20, 35, 34, 30, 21]
notas_2 = [22, 25, 32, 34, 12]

y = arange(len(legenda))
altura = 0.3
shift = (altura/2)
grafico_1 = plt.barh(y - shift, notas_1, height=altura, label="Notas P1", color="purple")
grafico_2 = plt.barh(y + shift, notas_2, height=altura, label="Notas P2", color="black")
plt.xlabel("Pontos")
plt.title("Pontuação por aluno")
plt.yticks(y, legenda)
plt.legend()
legenda_automatica(grafico_1)
legenda_automatica(grafico_2)
plt.show()
```

```python
legenda = ["Aluno1", "Aluno2", "Aluno3", "Aluno4", "Aluno5"]
notas_1 = [20, 35, 34, 30, 21]
notas_2 = [22, 25, 32, 34, 12]
notas_3 = [26, 34, 32, 33, 10]

y = arange(len(legenda))
altura = 0.3
shift = (altura/2)
grafico_1 = plt.barh(y - 3 * shift, notas_1, height=altura, label="Notas P1", color="purple", align="edge")
grafico_2 = plt.barh(y, notas_2, height=altura, label="Notas P2", color="black", align="center")
grafico_3 = plt.barh(y + shift, notas_3, height=altura, label="Notas P3", color="pink", align="edge")

plt.xlabel("Pontos")
plt.title("Pontuação por aluno")
plt.yticks(y, legenda)
plt.legend()
legenda_automatica(grafico_1)
legenda_automatica(grafico_2)
legenda_automatica(grafico_3)
plt.show()
```

### Utilizando decionário e função zip()

```python
alunos = ["Aluno1", "Aluno2", "Aluno3", "Aluno4", "Aluno5"]
notas_1 = [20, 35, 34, 30, 21]
notas_2 = [22, 25, 32, 32, 12]
notas_3 = [26, 34, 32, 33, 10]

dados_zip = zip(alunos, notas_1, notas_2, notas_3)
```

### A partir desse zip(), vamos criar um dicionário em que a chave seja o aluno e o valor seja a nota máxima obtida nas provas

```python
y = arange(len(alunos))
altura = 0.3

dados_dict = {dados[0]: max(dados[1], dados[2], dados[3]) for dados in dados_zip}
```

```python
def legenda_automatica(grafico):
    """ Adiciona a legenda acima da coluna do gráfico de acordo com a altura desta """
    for i in grafico:
        comprimento = i.get_width()
        plt.annotate(f'{comprimento}', xy=(comprimento, i.get_y() + i.get_height()/2), xytext=(10, 0), textcoords="offset points", ha="center")
```

```python
grafico1 = plt.barh(y, dados_dict.values(), height=altura, label="Nota máxima", color="purple", align="center")
plt.xlabel("Notas")
plt.title("Pontuação por aluno")
plt.yticks(y, dados_dict.keys())
plt.legend()
legenda_automatica(grafico1)
plt.show()
```
