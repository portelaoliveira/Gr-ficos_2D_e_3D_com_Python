```python
from matplotlib import pyplot as plt
from numpy import*
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 6 # Define as dimensões dos gráficos
```

```python
N = 10000000 # Número de dados gerados
n_colunas = 50 # Número de colunas do histograma

x = random.randn(N)
```

```python
#plt.hist(x, bins=n_colunas, density=False, histtype='bar', color='purple', label='Histograma x')
plt.hist(x, bins=n_colunas, density=True, histtype='bar', color='purple', label='Histograma x')
plt.legend()
plt.ylabel('N. de ocorrencias')
plt.title('Hostograma')
plt.show()
```

```python
N = 10000000 # Número de dados gerados
n_colunas = 20 # Número de colunas do histograma

x = random.randn(N, 4)
cores = ["orange", "blue", "green", "red"]
```

```python
#plt.hist(x, bins=n_colunas, density=False, histtype='bar', color='purple', label='Histograma x')
plt.hist(x, bins=n_colunas, density=True, histtype='bar', color=cores, label=cores)
plt.legend()
plt.ylabel('N. de ocorrencias')
plt.title('Hostograma')
plt.show()
```
