```python
from matplotlib import pyplot as plt
from numpy import*
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 6 # Define as dimensões dos gráficos
```

```python
legenda = ["Aluno1", "Aluno2", "Aluno3", "Aluno4", "Aluno5"]
notas = [20, 35, 34, 30, 21]

x = arange(len(legenda)) # Localização das legendas no eixo x
largura = 0.3 # Determina a largura das colunas
plt.bar(x, notas, width=largura, label="Notas P1", color="purple", align="edge")
plt.ylabel("Pontos")
plt.title("Pontuação por aluno")
plt.xticks(x, legenda)
plt.legend()
plt.show()
```

```python
def legenda_automatica(grafico):
    """ Adiciona a legenda acima da coluna do gráfico de acordo com a altura desta """
    for i in grafico:
        altura = i.get_height()
        plt.annotate(f'{altura}', xy=(i.get_x() + i.get_width()/2, altura), xytext=(0, 2), textcoords="offset points", ha="center")
```

```python
legenda = ["Aluno1", "Aluno2", "Aluno3", "Aluno4", "Aluno5"]
notas_1 = [20, 35, 34, 30, 21]
notas_2 = [22, 25, 32, 34, 12]

x = arange(len(legenda)) # Localização das legendas no eixo x
largura = 0.3 # Determina a largura das colunas
shift = (largura/2)
grafico1 = plt.bar(x - shift, notas_1, width=largura, label="Notas P1", color="purple")
grafico2 = plt.bar(x + shift, notas_2, width=largura, label="Notas P1", color="red")
plt.ylabel("Pontos")
plt.title("Pontuação por aluno")
plt.xticks(x, legenda)
plt.legend()
legenda_automatica(grafico1)
legenda_automatica(grafico2)
plt.show()
```
