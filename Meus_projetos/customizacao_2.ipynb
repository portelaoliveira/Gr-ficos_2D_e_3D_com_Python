```python
from matplotlib import pyplot as plt
from numpy import*
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 12, 10 # Define as dimensões dos gráficos
```

```python
t = arange(-5, 5, 0.01)
s = cos(2*pi*t)
plt.plot(t, s, color="pink", label="cos(2*$\pi$*t)")
plt.annotate("Valor máximo", xy=(2, 1), xytext=(3, 1.5), arrowprops=dict(facecolor="blue", shrink=0.02))
plt.legend()
plt.title("Função Cosseno")
plt.ylim(-2, 2)
plt.show()
```

```python
y = 2*t**2 - 3*t + 1
y_min = min(y)
x_min = 0

for i in zip(t, y):
    if i[1] == y_min:
        x_min = i[0]

plt.plot(t, y, color="purple", label="2° Grau")
plt.annotate("Ponto Mínimo", xy=(x_min, y_min), xytext=(x_min, y_min + 5), arrowprops=dict(facecolor="black", shrink=0.01))
plt.legend()
plt.title("Função 2° Grau")
plt.show()
```

```python
y1 = 2*t**2 + cos(t)
y_min = min(y1)
x_min = 0

for i in zip(t, y1):
    if i[1] == y_min:
        x_min = i[0]

plt.plot(t, y1, color="purple", label="2° Grau")
plt.annotate("Ponto Mínimo", xy=(x_min, y_min), xytext=(x_min, y_min + 5), arrowprops=dict(facecolor="black", shrink=0.01))
plt.legend()
plt.title("Função 2° Grau")
plt.show()
```
