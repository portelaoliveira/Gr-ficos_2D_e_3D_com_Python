```python
from matplotlib import pyplot as plt
from numpy import*
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 12, 10 # Define as dimensões dos gráficos
```

```python
t = arange(0.01, 10, 0.01)
dados1 = exp(t)
dados2 = sin(2*pi*t)
```

```python
plt.plot(t, dados1, color="red", label="exp(t)")
plt.yscale("log")
plt.xlabel("Tempo (s)")
plt.ylabel("exp(t)", color="red")
plt.legend()
plt.tick_params(axis="y", labelcolor="red")
plt.title("Gráficos em função do tempo")
plt.twinx()
plt.ylabel("sin(2*$\pi$*t)", color="purple")
plt.tick_params(axis="y", labelcolor="purple")
plt.plot(t, dados2, color="purple", label="sin(2*$\pi$*t)")
plt.legend()
plt.show()
```
