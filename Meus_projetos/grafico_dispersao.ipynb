```python
from matplotlib import pyplot as plt
```

### Uma função que recebe cinco parâmetros de entrada, sendo que os quatro últimos são opcionais
### com default igual a 1.

```python
def funcao_terceiro_grau (x, a=1, b=1, c=1, d=1):
    y = a*x**3 + b*x**2 + c*x + d
    return y

def funcao_seg_grau (x, a=1, b=1, c=1):
    y = a*x**2 + b*x + c
    return y
```

```python
lista_x = range(-5, 5)
lista_y_3 = []
lista_y_2 = []

for i in lista_x:
    lista_y_3.append(funcao_terceiro_grau(i))
    lista_y_2.append(funcao_seg_grau(i, b=-3))
    
#plt.scatter(lista_x, lista_y)
plt.plot(lista_x, lista_y_3, label="3 garu", color="gray")
plt.plot(lista_x, lista_y_2, label="2 garu", color="purple")
plt.legend()
plt.xlabel("Valores de x")
plt.ylabel("Valores de y")
plt.title("Função do terceiro grau")
plt.show()
```

```python
lista_map = list(map(funcao_seg_grau, lista_x))
plt.plot(lista_x, lista_map, label="2 grau", color="green")
plt.legend()
plt.xlabel("Valores de x")
plt.ylabel("Valores de y")
plt.title("Função do segundo grau")
plt.show()
```
